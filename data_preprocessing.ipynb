```python
df = df_adop[['Id', 'Pawpularity']]
df
```

```python
import os
path = 'adoption/train_images'
files = os.listdir(path)
data = []
for index, file in enumerate(files):
    name = file.split('.')[0].split('-')[0]
    score = df[df['Id'] == name]['Pawpularity'].values[0]
    # print(name, file, score)
    data.append([file, score])
```

```python
df_new = pd.DataFrame(data, columns=['Id', 'Pawpularity'])
df_new
```

```python
df_new.to_csv('train_new.csv', sep=str(','), index=False)
```
